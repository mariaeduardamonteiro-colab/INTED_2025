```python
#3 – Um sistema meteorológico quer classificar a sensação térmica do dia. A partir de uma temperatura informada, determine se está fria, agradável, quente ou muito quente usando if…elif…else.
temperatura = float(input("digite a temperatura"))
if temperatura < 15:
  print("sensação térmica: fria")
elif 15 <= temperatura < 25:
  print("sensação trémica: agradável")
elif 25 <= temperatura < 35:
  print("sensção térmica: quente")
else:
  print("sensação térmica: muito quente")
```

```python
#5 – Uma empresa deseja reajustar salários com base em faixas de rendimento. Receba o salário atual e aplique um aumento conforme as faixas (10%, 7% ou 5%), usando condicionais encadeadas. Exiba o novo salário.
salário_atual = float(input("digite o salário atual:"))
if salário_atual <= 2000:
  percentual_aumento = 0.10 # 10%
elif salário_atual <= 5000:
  percentual_aumento = 0.7 # 7%
else:
  percentual_aumento = 0.5 # 5%

novo_salário = salário_atual * (1 + percentual_aumento)

print(f"salário atual: R${salário_atual:.2f}")
print(f"percentual de aumento: {percentual_aumento * 100:.0f}%")
print(f"novo salário: R${novo_salário:.2f}")
```

```python
#14 – Uma pesquisa está registrando idades. Continue lendo idades até que o usuário digite um valor negativo. Ao final, calcule e exiba a média das idades digitadas.
soma_idades = 0
quantidade_idades = 0
print("digite as idade (e um num negativo para encerrar):")
while True:
  idade = int(input("idade:  "))
  if idade < 0:
    break
  soma_idades += idade
  quantidade_idades += 1
if quantidade_idades > 0:
  média = soma_idades / quantidade_idades
  print(f"a média das idades é: {média:.2f}")
else:
  print("nenhuma idade válida para calculara média:")
```

```python
#15 – Um cronômetro digital deve realizar uma contagem regressiva personalizada. Receba um número inicial e exiba a contagem até 0 usando um while.

num_inicial = int(input("digite o num inicial:"))
print(f"iniciando contagem regressiva {num_inicial}")
while num_inicial >= 0:
  print(num_inicial)
  num_inicial -= 1
print("contagem finalizada")
```

```python
#21 – Um app de produtividade quer identificar qual de quatro tarefas possui o nome mais longo. Leia quatro nomes de tarefas e exiba aquela com mais caracteres.

tarefa_mais_longa = input("Digite o nome da 1ª tarefa: ")

for i in range(2, 5):
    tarefa = input(f"Digite o nome da {i}ª tarefa: ")
    if len(tarefa) > len(tarefa_mais_longa):
        tarefa_mais_longa = tarefa

print(f"A tarefa com o nome mais longo é: {tarefa_mais_longa}")
```

```python
#23 – Um módulo de análise precisa classificar uma lista de valores. Leia 10 números e exiba quantos são positivos, negativos e quantos são iguais a zero.

positivos = 0
negativos = 0
zeros = 0

for i in range(10):
    num = float(input(f"Digite o {i+1}º número: "))

    if num > 0:
        positivos += 1
    elif num < 0:
        negativos += 1
    else:
        zeros += 1

print(f"Positivos: {positivos}")
print(f"Negativos: {negativos}")
print(f"Zeros: {zeros}")
```

```python
#27 – Crie um sistema de login com limite de tentativas. O usuário terá no máximo 3 tentativas para digitar a senha correta. Caso acerte antes, use break para encerrar o laço.

senha_correta = "12345"

for tentativa in range(1, 4):
    senha = input(f"Tentativa {tentativa} de 3: Digite a senha: ")

    if senha == senha_correta:
        print("Senha correta")
        break
    else:
        print("Senha incorreta.")
else:
    print("Número máximo de tentativas atingido")
```

```python
#28 – Desenvolva um programa que leia 10 números e some apenas os valores positivos. Caso um número negativo seja digitado, utilize continue para ignorá-lo.

soma = 0

for i in range(10):
    num = float(input(f"Digite o {i+1}º número: "))

    if num < 0:
        continue

    soma += num

print(f"Soma dos números positivos: {soma}")
```

```python
#29 – Simule um acumulador que percorre números até que a soma ultrapasse 200. Quando isso ocorrer, encerre o laço usando break.

soma = 0

while True:
    num = float(input("Digite um número: "))
    soma += num

    if soma > 200:
        print("A soma ultrapassou 200!")
        break

print(f"Soma final: {soma}")
```
